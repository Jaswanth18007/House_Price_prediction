```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
df=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
df.shape#formula to remove outliers
```

```python
fig,axs=plt.subplots(2,3,figsize=(10,5))
plt1=sns.boxplot(x=df['price'],ax=axs[0,0])
plt2=sns.boxplot(x=df['area'],ax=axs[0,1])
plt3=sns.boxplot(x=df['bedrooms'],ax=axs[0,2])
plt1=sns.boxplot(x=df['bathrooms'],ax=axs[1,0])
plt2=sns.boxplot(x=df['stories'],ax=axs[1,1])
plt3=sns.boxplot(x=df['parking'],ax=axs[1,2])
plt.tight_layout()
#all outliers are removed
```

```python
#Data preparation
```

```python
#spliting of dataset
x=df.iloc[:,1:]#others are (independent values)
y=df.iloc[:,0]#price values(dependent values)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()#converting categorical values to numeric values
```

```python
x[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']]=oe.fit_transform(x[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']])
```

```python
print(x[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']])
```

```python
x[['furnishingstatus']]=oe.fit_transform(x[['furnishingstatus']])#converting categorical values to numeric values
print(x[['furnishingstatus']])
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
from sklearn.model_selection import train_test_split#testing and training the model
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
model.intercept_#m value
```

```python
model.coef_#c value
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(ytest,ypred))
print(mean_squared_error(ytest,ypred))
```

```python
print(model.score(xtrain,ytrain))
print(model.score(xtest,ytest))
print(r2_score(ytest,ypred))
```
